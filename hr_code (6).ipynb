## 3) CREATE AND EVALUATE THE MODEL
I created a ***Logistical Regression Model*** with the data to predict attrition. 

I then evaluated how well it did and tried to make it as effective as possible. To prevent overfitting, I trimmed down the model to use the least amount of variables. I did this by eliminating variables based on it's effectiveness with the ***Area Under the Curve*** chart. I tried using the least amount of variables so that I could give specific recommendations rather than a broad recommendation to focus on everything.  

I split the data up to a test set and a train set. I used the training data to make a model, I then ran the model onto the training data. This allowed me to see  how well it works using a ***Area Under the Curve*** ROC method. 

This section is composed of two parts:

1- Cleaning up the data

2- Creating a *Logistical Regregression* prediction model


### 1- CLEAN UP THE DATA 
Here I took the data and split it to be used in the two models.

```r
#Check for null values and omit any that are null
sum(is.na(hr_data))
```

```r
#Split the data into two sets. A training data set to build a model and a test set to use the model on. 
set.seed(3456)
trainIndex <- createDataPartition(hr_data$Attrition, p = .8, 
                                  list = FALSE, times = 1)
hr_train <- hr_data[trainIndex,]
hr_test <- hr_data [-trainIndex,]
```

```r
#use head fxn to look at the data that has been split.
head(hr_train)
head(hr_test)
```

### 2- Logistic Regression Model
***I first started with all variables as a function of attrition.*** 

Then I trimmmed the data of all variables that are not statistically significant. This allowed me to mold the data into a model to predict if an employee quits. I also used variable importance plot to trim down even more. 

I ended up getting the most effective model with *Logistic Regression Model 2*.

This is what I ultimately used to make predictions. 

I used the Area Under the Curve graphs to see the effectiveness of the models as I trimmed the variables. 


```r
#use glm to create a model to predict attrition. Then check their P values using summary fxn
attrition_prediction_lr_1 <- glm(Attrition ~ ., data = hr_train)
summary(attrition_prediction_lr_1)
```

```r
#use this model to predict attrition on the test data. Use this to evaluate how effective the model is. 
hr_test$Attrition_Predictions_lr_1 <- predict(attrition_prediction_lr_1, newdata = hr_test, type = "response")
head(hr_test)
```

```r
#evaluate variables 
#Take out unimportant variables. 
varImp(attrition_prediction_lr_1)
```

```r
attrition_prediction_lr_2 <- glm(Attrition ~ Age + OverTime + 
                                    JobInvolvement + YearsInCurrentRole, data = hr_train)
summary(attrition_prediction_lr_2)
```

```r
#use this model to predict attrition on the test data. Use this to evaluate how effective the model is. 
hr_test$Attrition_Predictions_lr_2 <- predict(attrition_prediction_lr_2, newdata = hr_test, type = "response")
head(hr_test)
```

```r
varImp(attrition_prediction_lr_2)
```

```r
# Evaluate the logistic regression models.

# Generate ROC curves for both version of the L.R. model.
ROC_lr_1 <- roc(hr_test$Attrition, hr_test$Attrition_Predictions_lr_1)
ROC_lr_2 <- roc(hr_test$Attrition, hr_test$Attrition_Predictions_lr_2)

# Print the AUC for both versions of the model
paste("AUC for Logistic Regression Model (all variables):", round(auc(ROC_lr_1),2), "(red)")
paste("AUC for Logistic Regression Model (limited variables):", round(auc(ROC_lr_2),2), "(blue)")

# Plot the ROC curves.
plot.roc(ROC_lr_1, col="red")
lines.roc(ROC_lr_2, col="blue")
```

This graph shows that even though we narrowed the variables down to only 4 variables, the area under the curve only decreased by a small amount. 

The reason I was willing to take a slighlty less effective model is that I can better apply this to future data sets and I can make specific recommendations based on a few variables. This is better than recommending the company to focus on every single variable provided.
